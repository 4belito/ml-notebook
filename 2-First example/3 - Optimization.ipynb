```python
import matplotlib.pyplot as plt
import torch
```

```python
def f(x: torch.Tensor) -> torch.Tensor:
    return 3*x**2 + torch.sin(5*x)

x = torch.linspace(-1, 1, 100)

plt.plot(x, f(x))
plt.title('Function to minimize')
plt.show()
```

## Gradient descent

```python
def grad_f(x: torch.Tensor) -> torch.Tensor:
    return 6*x + 5*torch.cos(5*x)
```

```python
xk = torch.randn(1)
N_epoachs = 20
lr = 0.01
all_x = [xk.item()]
for _ in range(N_epoachs):
    xk = xk - lr * grad_f(xk)
    all_x.append(xk.item())
    
all_x=torch.tensor(all_x)
```

```python
plt.plot(x,f(x), label='f(x)')
plt.plot(all_x,f(all_x), 'o--', label='xk')
plt.title('Gradient descent')
plt.show()
```
