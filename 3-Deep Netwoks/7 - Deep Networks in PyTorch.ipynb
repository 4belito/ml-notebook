```python
import torch
import torch.nn as nn
```

```python
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(1, 10)
        self.fc2 = nn.Linear(10, 1)
        self.act = nn.ReLU()

    def forward(self, x):
        x = self.act(self.fc1(x))
        x = self.fc2(x)
        return x
```

```python
net = Net()
for i,param in enumerate(net.parameters()):  # Example to show parameters
    print(f' net patram {i} : {param.grad}')
r = net(torch.tensor([1.0]))  # Example forward pass
r.backward()
for i,param in enumerate(net.parameters()):  # Example to show parameters
    print(f' net patram {i} : {param.grad}')
```
