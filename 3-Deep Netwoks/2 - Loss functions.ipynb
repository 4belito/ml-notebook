```python
import torch
import torch.nn as nn
```

## Regression

In practice does not matter too much

### L1 
$$
\text{L1}(x,y)=|x-y|
$$

```python
nn_l1 = nn.L1Loss(reduction="mean") 

def L1Loss():
    def l1(x,y):
        return torch.mean(torch.abs(x - y))
    return l1

l1 = L1Loss()

x = torch.tensor([2.0, 3.0, 4.0])
y = torch.tensor([1.0, 2.0, 2.0])

nnl1_xy=nn_l1(x, y).item()
l1_xy = l1(x, y).item()
print(f'{nnl1_xy=}') 
print(f'{l1_xy=}')
```

### L2 (MSE)

$$
\text{L2}(x,y) = (x-y)^2
$$

```python
nn_mse = nn.MSELoss(reduction="mean") 

def MSELoss():
    def l2(x,y):
        return torch.mean((x - y)**2)
    return l2

mse=MSELoss()

x = torch.tensor([2.0, 3.0, 4.0])
y = torch.tensor([1.0, 2.0, 2.0])

nnmse_xy=nn_mse(x, y).item()
mse_xy = mse(x, y).item()
print(f'{nnmse_xy=}') 
print(f'{mse_xy=}')
```
