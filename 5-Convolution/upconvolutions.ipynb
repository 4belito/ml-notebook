```python
import numpy as np
import torch
```

## Upconvolution

```python
def upconv(A,F,stride=(1,1)):
    n,m=A.shape
    result=[]
    s1,s2=stride
    for i in range(A.shape[0]):
        for j in range(A.shape[1]):
            multiply=A[i,j]*F
            padded=np.pad(multiply, ((i*s1, (n-1-i)*s1), (j*s2, (m-1-j)*s2))) 
            result.append(padded)
    return np.sum(result,axis=0)

def upconv_torch(A,F,stride=(1,1)):
    upconv_torch = torch.nn.ConvTranspose2d(1, 1, kernel_size=F.shape, 
                                            stride=stride, bias=False)
    upconv_torch.weight = torch.nn.Parameter(
        torch.tensor(F).unsqueeze(0).unsqueeze(0).float()
        )
    A_torch = torch.tensor(A).unsqueeze(0).unsqueeze(0).float()
    return upconv_torch(A_torch).detach().squeeze().numpy()
```

```python
A = np.ones((4,4))
F = np.ones((3,3))
print(f"{A=}\n{F=}")
```

```python
output_np = upconv(A,F,stride=(1,1))
print(f"{output_np=}")
```

```python
output_torch = upconv_torch(A,F,stride=(1,1))
print(f"{output_torch=}")
```
