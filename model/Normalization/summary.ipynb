# Normalizations

# Batch Normalization

* ✅ Regularizes the network (avoiding activation vanishing and exploding)
* ✅ Handles badly scaled weights 
* ❌ Mixes gradient info between batches
* ❌ It is sligthly annoying to use because discrepancy between training and testing and also for multiple CPUs because it requires the mean and std to be sync across the entire batch

**Note:** 
* Large batch sizes work better becase batch mean and std are closer to data mean and std
* There is not mean and std at test-time, so batch normalization is implemented via running average

# Layer Normalization

* ✅ Training and testing are the same
* ✅ Works well for sequence model
* ❌ Does not normalize activations individually
